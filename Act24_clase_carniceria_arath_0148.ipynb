### Clase carniceria
En este programa, la clase Carniceria actúa como un molde o plantilla que define las características y comportamientos de cualquier negocio de este tipo. Al crear dos objetos diferentes, estamos realizando un proceso llamado instanciación.

Cada objeto (o instancia) es independiente: aunque comparten las mismas funciones (cortar, empacar, pesar), cada uno mantiene sus propios datos de temperatura, proveedores y nombres sin mezclarse entre sí. Esto permite que el código sea organizado, escalable y fácil de mantener.
![texto del vínculo](https://raw.githubusercontent.com/JacoboValenzuela128/Act24-clase-carniceria-arath-0148/refs/heads/main/infoclasecarniceria.png)

```python
class Carniceria:
    # El constructor define los atributos
    def __init__(self, nombre, tipo_carnes, proveedores, temperatura_camara):
        self.nombre = nombre
        self.tipo_carnes = tipo_carnes
        self.proveedores = proveedores
        self.temperatura_camara = temperatura_camara

    # Funciones (Métodos)
    def cortar_carne(self):
        print(f"🔪 En '{self.nombre}' están cortando {self.tipo_carnes}.")

    def empacar(self):
        print(f"📦 Empacando productos de {self.proveedores}.")

    def pesar_producto(self):
        print(f"⚖️ Pesando producto. Refri a: {self.temperatura_camara}°C.")

# --- Registro del Primer Objeto ---
print("🥩 REGISTRO DE LA CARNICERÍA #1")
nom1 = input("Nombre: ")
tip1 = input("Tipos de carne: ")
pro1 = input("Proveedores: ")
tem1 = input("Temperatura (°C): ")
carniceria1 = Carniceria(nom1, tip1, pro1, tem1)

print("\n" + "-"*30 + "\n")

# --- Registro del Segundo Objeto ---
print("🥩 REGISTRO DE LA CARNICERÍA #2")
nom2 = input("Nombre: ")
tip2 = input("Tipos de carne: ")
pro2 = input("Proveedores: ")
tem2 = input("Temperatura (°C): ")
carniceria2 = Carniceria(nom2, tip2, pro2, tem2)

# --- Ejecución de acciones para ambos ---

print("\n=== REPORTE FINAL DE ACTIVIDADES ===")

print(f"\n--- Actividad de {carniceria1.nombre} ---")
carniceria1.cortar_carne()
carniceria1.empacar()
carniceria1.pesar_producto()

print(f"\n--- Actividad de {carniceria2.nombre} ---")
carniceria2.cortar_carne()
carniceria2.empacar()
carniceria2.pesar_producto()

print("\n-------------------------------------------------------------------")
print(" Programa interactivo creado por Arath Valenzuela (NL52) & Gemini AI")
```
